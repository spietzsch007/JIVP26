# Predictions

Load a model with a .pt file.  
(The .pt files could be found in the results of an training)

```python
model = YOLO('../../models/yolov8seg/YOLOv8RCF.pt')
```

Create a list with file paths to the images.

```python
predict_paths = glob.glob('../../test_images/rcf/*.jpg')
```

Find Segmentation Masks in your images and save them.  
Change the device to 'cuda' or '0' for the first cuda device to run on GPU, if you have an compatible card.  
Note: use timeit to measure the time independed of YOLO. For example with 10 runs and 1 loop: %%timeit -r 10 -n 1.

```python
#%%timeit -r 10 -n 1
predictions = model(predict_paths, conf = 0.4, save = False, save_txt = False, max_det = 1, classes = [0], show_labels = False, name = 'predictiontest', device = 'cpu')
```
