## YOLO Segmentation Prediction

The workflow includes:
1. Import libraries
2. Loading the trained YOLOv8 model (.pt file)
3. Loading input images from the `test_images/` folder
4. Running prediction on all test-images
5. Saving the prediction outputs as text files

```python
import glob
from ultralytics import YOLO
import general_functions as gf
```

```python
model= YOLO('.../model/YOLOv8Damage.pt')
```

```python
predict_paths = glob.glob('.../test_images/surface_damage/*.png')
```

```python
predictions = model(predict_paths, conf = 0.5, save = False, save_txt = False, max_det = 3, classes = [0], show_labels = False, name = 'predict01')
```

### If several objects are detected in the image, save only the outline of the central object in a text file.
->> Only text files with a single contour are required for further processing <<-


```python
save_path='.../outputs/yolov8seg/surface_damage/'
save_masks=gf.find_central_mask_save_as_yolo_txt(predictions,save_path)
```
